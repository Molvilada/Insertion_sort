```python
from itertools import *
```

```python
def insertion_short(A):

    comparaciones = 0
    intercambios = 0
    instrucciones = 0
    for i in range(1,len(A)):
        temp = A[i]
        j = i - 1
        while j > -1 and temp < A[j]:
            comparaciones += 1
            A[j+1] = A[j]
            j = j-1
            intercambios += 1
            instrucciones += 3
        instrucciones += 1
        A[j+1] = temp
        instrucciones += 4

        if j != i:
            comparaciones += 1

    instrucciones += 1
    resultados = []
    resultados.append(instrucciones)
    resultados.append(comparaciones)
    resultados.append(intercambios)
    return resultados
```

```python
for n in range(2,11):
    A = list(i for i in range(n))
    acum_ins = 0
    acum_c = 0
    acum_int = 0
    long = 0
    for a in permutations(A):
        acum = insertion_short(list(a))
        acum_ins += acum[0]
        acum_c += acum[1]
        acum_int += acum[2]
        long += 1

    print('ARREGLO DE TAMAÑO {}'.format(n))
    print('La cantidad de instrucciones promedio es: {}'.format(float(acum_ins / long)))
    print('La cantidad de comparaciones promedio es: {}'.format(float(acum_c / long)))
    print('La cantidad de intercambios promedio es: {}'.format(float(acum_int / long)))
    print('-----------------------')
```
